```python
import os
import numpy as np
```

```python
path_to_gt = "/content/jhu_crowd_v2.0/test/gt/"
```

```python
gts = dict()
np.random.seed(0)
```

```python
for fn in os.listdir(path_to_gt):

  fn_jpg = fn.replace(".txt", ".jpg")

  arr = np.loadtxt(path_to_gt + fn)
  if len(arr) == 0 or arr.ndim != 2:
    continue

  sample_size = len(arr) if len(arr) < 3 else 3

  gt = {}
  gt["points"] = arr[:, :2].tolist()

  box_examples_coordinates = []
  idx = np.random.choice(range(len(gt["points"])), size = sample_size, replace = False)
  for i in idx:
    row = arr[i].astype(int)

    x = row[0]
    y = row[1]
    
    half_width = row[2]
    half_height = row[3]
    ul = [x - half_width, y - half_height]
    ur = [x + half_width, y - half_height]
    lr = [x + half_width, y + half_height]
    ll = [x - half_width, y + half_height]
    box_examples_coordinates.append([ul, ur, lr, ll])

  gt["box_examples_coordinates"] = box_examples_coordinates
  gts[fn_jpg] = gt
```

```python
import json

def convert(o):
    if isinstance(o, np.generic): return o.item()  
    raise TypeError

with open('annotation_JHU_Crowd.json', 'w') as fp:
  json.dump(gts, fp, default=convert)
```

```python
splits = {}
splits["test"] = [fn.replace(".txt", ".jpg") for fn in os.listdir(path_to_gt)]

with open('Train_Test_Val_JHU_Crowd.json', 'w') as fp:
  json.dump(splits, fp)
```
